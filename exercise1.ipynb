```python
def string_ops(s):
    return s[::-1], s[::2]
```

```python
nums = [1,2,3,4,5,6]
squares = [x**2 for x in nums if x % 2 == 0]
```

```python
def freq(s):
    d = {}
    for ch in s:
        d[ch] = d.get(ch, 0) + 1
    return d
```

```python
def odd_indices(lst):
    return [i for i, x in enumerate(lst) if x % 2 != 0]
```

```python
def rotate(lst, k):
    return lst[-k:] + lst[:-k]
```

```python
def make_dict(a, b):
    return dict(zip(a, b))
```

```python
lst = [1,2,3]
it = iter(lst)

while True:
    try:
        print(next(it))
    except StopIteration:
        break
```

```python
def flatten(lst):
    return [x for sub in lst for x in sub]
```

```python
def palindrome(s):
    return s == s[::-1]
```

```python
def second_largest(lst):
    first = second = float('-inf')
    for x in lst:
        if x > first:
            second = first
            first = x
        elif x > second and x != first:
            second = x
    return second
```
